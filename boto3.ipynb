```python
import boto3
import requests
# Crea un client S3 compatibile puntatndo al server minIO

s3 = boto3.client(
    "s3",
    endpoint_url="http://localhost:9000",
    aws_access_key_id="admin",
    aws_secret_access_key="admin123",
)
```

```python
#Carica un file su minIO
s3.upload_file(
    Filename="mat.jpg",
    Bucket="nuovo",
    Key="Mattarella"
)
print ("upload completato")
```

```python
#Scarica un file su minIO
s3.download_file(
    Filename="mat2.jpg",
    Bucket="nuovo",
    Key="Mattarella"
)
print ("download completato")
```

```python
# elenca file nel bucket

response = s3.list_objects_v2(Bucket="nuovo") #V2 metodo usato per elencare file

for obj in response.get("Contents",[]):
    print(obj["Key"])
```

```python
# cancella file dal bucket
s3.delete_object(
    Bucket="nuovo",
    Ket="Mattarella"
)
print("delete completato")
```

```python
# Crea bucket
s3.create_bucket(Bucket="nuovo-bucket")
print("bucket created")
```

```python
s3.delete_bucket(
    Bucket="nuovo-bucket"
)
```
